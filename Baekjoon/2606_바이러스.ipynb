### 1차 시도 : 실패

```python
import sys
input= sys.stdin.readline
# sys.setrecursionlimit(10**6) # 이걸 안 하면 런타임에러(RecursionError) 해도 시간초과, 최소 에러는 해소된다.

n= int(input().rstrip())
m= int(input().rstrip())

box= {i+1: [] for i in range(n)}
for i in range(m):
    k,v = map(int, input().rstrip().split())
    box[k].append(v)

def sol(network, first, next_num= [0], cnt= set()):
    for i in network[first]:
        cnt.add(i)
        next_num.insert(0, i)
    
    next_pop= next_num.pop(0)
    if next_pop == 0:
        return len(cnt)
    else:
        return sol(network, next_pop, next_num, cnt)
    
sol(box, 1)
```

### 6차 시도 : 성공

```python
'''
예제를 아래와 같은 형태가 아닌
{1:[2,5], 2:[3], 3:[], 4:[7], 5:[2,6], 6:[], 7:[]}
아래와 같은 형태의 딕셔너리로 담는다.
{1:[2,5], 2:[3,1,5], 3:[2], 4:[7], 5:[2,6,1], 6:[5], 7:[4]}

'''

from sys import stdin
input= stdin.readline

n= int(input().rstrip()) # 7
m= int(input().rstrip()) # 6

box= {i+1: [] for i in range(n)} # box = {1:[], 2:[]...7:[]}
for i in range(m):
    k,v= map(int, input().rstrip().split())
    box[k].append(v)
    box[v].append(k)
    # box = {1:[2,5], 2:[3,1,5], 3:[2], 4:[7], 5:[2,6,1], 6:[5], 7:[4]}

def sol(k, after= [], before= [], cnt= 0):
    before.append(k) # 검사 끝낸 딕셔너리 인덱스
    '''
        <1차>
        k = 1
        before = [1]
    '''

    for i in box[k]:
        after.append(i) # 검사 할 값
        '''
            <1차>
            after = [2,5] => box[1]값 추가
        '''

    while after != []: # after에 값이 없을 때까지 반복
        n= after.pop(0)
        '''
            <1차>
            n = 2
            after = [5]
        '''
        if n not in before: # 검사한 값에 검사할 값이 없다면
            cnt += 1
            return sol(n, after, before, cnt) # 재귀
            '''
                <1차>
                sol(2, [5], [1], 1)
                    before = [1, 2]
                    after = [5, 3, 1, 5] => box[2] 추가
                    n = 5, after = [3,1,5]
                    sol(5, [3,1,5], [1, 2], 2)
                        before = [1,2,5]
                        after = [3,1,5,2,6,1] => box[5] 추가
                        n = 3, after = [1,5,2,6,1]
                        sol(3, [1,5,2,6,1], [1,2,5], 3)
                            before = [1,2,5,3]
                            after = [1,5,2,6,1,2] => box[3] 추가
                            n = 1, after = [5,2,6,1,2]
                            n = 5, after = [2,6,1,2]
                            n = 2, after = [6,1,2]
                            n = 6, after = [1,2]
                            sol(6, [1,2], [1,2,5,3], 4)
                                befoer= [1,2,5,3,6]
                                after = [1,2,5] => box[5] 추가
                                n = 1 after = [2,5]
                                n = 2 after = [5]
                                n = 5 after = []
                                after가 비었으니까 while을 나간다.(재귀도 드디어 끝)

            '''
        
    return cnt # cnt = 4

print(sol(1)) # 1과 연결된 값을 찾는 거니까
```

### 다른 사람 코드
내 코드랑 비교하기 위해 변수같은 것들 내 코드랑 맞춤

```python
from sys import stdin
input= stdin.readline

n= int(input()) # 7
m= int(input()) # 6
box= []
cnt= set()

# 1과 직접 연결된 숫자를 찾아서 cnt에 넣는다.
for _ in range(m):
    k, v = map(int, input().split())
    if k == 1 or v == 1: # k,v 중에 1이 있다면
        cnt.add(k) # cnt 집합에 k 추가
        cnt.add(v) # cnt 집합에 v 추가
    else: # k,v 중에 1이 없다면
        box.append([k,v]) # box에 k,v를 리스트 형태로 추가
        '''
            1/6 번째
            k = 1, v = 2
            cnt = set(1,2)
            box = []

            2/6 번째
            k = 2, v = 3
            cnt = set(1,2)
            box = [[2,3]]

            3/6 번째
            k = 1, v = 5
            cnt = set(1,2,5)
            box = [[2,3]]

            4/6 번째
            k = 5, v = 2
            cnt = set(1,2,5)
            box = [[2,3], [5,2]]

            5/6 번째
            k= 5, v = 6
            cnt = set(1,2,5)
            box = [[2,3], [5,2], [5,6]]

            6/6 번째
            k = 4, v = 7
            cnt = set(1,2,5)
            box = [[2,3], [5,2], [5,6], [4,7]]

            cnt에는 1과 직접 연결된 모든 번호가 담겨있다.
            box에는 1과 직접 연결되지 않은 번호가 담겨있다.
            
        '''

# 1과 직접 연결되지는 않고 건너서 연결되는 값을 cnt에 넣는다.
while True:
    # cnt = set(1,2,5)
    # box = [[2,3], [5,2], [5,6], [4,7]]
    for i in box:
        if i[0] in cnt:
            cnt.add(i[1]) 
            box.remove(i) # 삭제 안하면 box에 값이 계속 있는 거니까 무한 반복함
            break # 삭제되 box 리스트로 다시 for문 돌아감
        elif i[1] in cnt:
            cnt.add(i[0])
            box.remove(i) # 삭제 안하면 box에 값이 계속 있는 거니까 무한 반복함
            break # 삭제되 box 리스트로 다시 for문 돌아감

        '''
            i = [2,3]
            cnt = set(1,2,5,3) 3 추가
            box = [[5,2], [5,6], [4,7]]
            2는i[0]는 1과 연결되어 있고 3과 연결되어 있으니까 cnt에 넣는다는 뜻

            i = [5,2]
            cnt = set(1,2,5,3) # 이미 값이 있어 추가 없음
            box = [[5,6], [4,7]]

            i = [5,6]
            cnt = set(1,2,5,3,6) # 6 추가
            box = [[4,7]]

            i = [4,7]
            cnt = set(1,2,5,3,6)
            box = []

        '''
    
    # for-else 문으로 for문이 중간에 break로 끝나지 않고 끝까지 수행되었을 때 수행된다.
    else:
        break

# cnt = set(1,2,5,3,6)
print(len(cnt) - 1) # 1은 제외하고 개수 출력
```
