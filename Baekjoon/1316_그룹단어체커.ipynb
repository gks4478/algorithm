## 1316 그룹 단어 체커
https://www.acmicpc.net/problem/1316

- 각 문자가 연속해서 나타나는 경우만 카운트한다
- 연속하다가 떨어져서 나타나면 카운트하지 않는다

<예시>

```python
from sys import stdin
input= stdin.readline

result= 0

for _ in range(int(input())):
    input_str= input().rstrip()
    v= 0 # 연속된 단어가 떨어져 있는 경우는 1이 된다

    # 단어가 하나거나 전체 단어 개수가 1개면 result에 1을 더한다
    if len(set(input_str)) == 1:
        result += 1

    else:
        for i in range(len(input_str)):
            for j in range(i+2, len(input_str)):
                # i번째 단어와 i+2번째 단어가 같고
                # i번째 단어와 i+1번째 단어가 다르면 v는 1이 된다
                if input_str[i] == input_str[j] and input_str[i] != input_str[j-1]:
                    v = 1
                    break
                
        # v가 1이 아니면 result에 1을 더한다
        if v == 0:
            result += 1

print(result)
```
