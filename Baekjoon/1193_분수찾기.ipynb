### python으로는 통과 못하고 pypy3로만 통과함

```python
from sys import stdin
input= stdin.readline

n= int(input())

cnt= 0
for i in range(1, n+1):
    for _ in range(i):
        cnt += 1
        if cnt == n:
            a= [x for x in range(i)]
            break
    if cnt == n:
        break

point= ((i-1) * i) // 2

if i%2 == 0:
    for x,y in zip(a, a[::-1]):
        point += 1
        if point == n:
            print(str(x+1) + '/' + str(y+1))
else:
    for x,y in zip(a[::-1], a):
        point += 1
        if point == n:
            print(str(x+1) + '/' + str(y+1))
```
