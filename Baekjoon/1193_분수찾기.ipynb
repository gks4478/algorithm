- i가 짝수면 a에 오름차순으로 더해지고 홀수면 b에 내림차순으로 더해진다
- i가 홀수면 a에 내림차순으로 더해지고 짝수면 b에 오름차순으로 더해진다

### python으로는 통과 못하고 pypy3로만 통과함

```python
from sys import stdin
input= stdin.readline

n= int(input())

# a에 각 더해지는 값을 추가한다(i=2에 [0,1,2]처럼)
cnt= 0
for i in range(1, n+1):
    for _ in range(i):
        cnt += 1
        if cnt == n:
            a= [x for x in range(i)]
            break
    if cnt == n:
        break

point= ((i-1) * i) // 2

# i가 짝수면 x가 오름차순(a), y가 내림차순(a[::-1])
if i%2 == 0:
    for x,y in zip(a, a[::-1]):
        point += 1
        if point == n:
            print(str(x+1) + '/' + str(y+1))
    
# i가 홀수면 x가 내림차순(a[::-1]), y가 오름차순(a)
else:
    for x,y in zip(a[::-1], a):
        point += 1
        if point == n:
            print(str(x+1) + '/' + str(y+1))
```
