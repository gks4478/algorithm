## 9095 1,2,3 더하기
https://www.acmicpc.net/problem/2839

정수 4를 1,2,3의 합으로 나타내는 방법은 총 7가지가 있다.
- 1+1+1+1
- 1+1+2
- 1+2+1
- 2+1+1
- 2+2
- 1+3
- 3+1

정수 n이 주어졌을 때 n을 1,2,3의 합으로 나타내는 방법의 수를 구하라

<예시>

<pre>
[입력]
3
4
7
10

[출력]
7
44
274
     
</pre>

1 -> 1개
- 1

2 -> 2개
- 1 + 1
- 2

3 -> 3개 
- 1 + 1 + 1
- 1 + 2
- 2 + 1
- 3

4 -> 7개
- 1 + 1 + 1 + 1
- 1 + 1 + 2
- 1 + 2 + 1
- 2 + 1 + 1
- 2 + 2
- 3 + 1
- 1 + 3

5 -> 13개
- 1 + 1 + 1 + 1 + 1
- 1 + 1 + 1 + 2
- 1 + 1 + 2 + 1
- 1 + 2 + 1 + 1
- 2 + 1 + 1 + 1
- 2 + 2 + 1
- 2 + 1 + 2
- 1 + 2 + 2
- 3 + 1 + 1
- 1 + 3 + 1
- 1 + 1 + 3
- 3 + 2
- 2 + 3

...

규칙이 3부터는 안 보이고 4부터 보인다.
- 4가 7개의 방법으로 구할 수 있는데 이는 앞 1, 2, 3의 방법의 수를 합친것과 같다.
- => 4: 1 + 2 + 4 = 7
- 5가 13개의 방법으로 구할 수 있는데 이는 앞 2, 3, 4의 방법의 수를 합친것과 같다.
- => 5: 2 + 4 + 7 = 13

```python
def sol(n):
    # 이 부분이 없으면 그냥 0,1,2,3을 입력하면
    # 아래 리스트에 직접 삽입하는 부분에서 오류가 난다.(a[1]=1; a[2]=2; a[3]=4 부분)
    if n < 3:
        return n
        # n이 0,1,2 중에 있다면 그대로 출력한다.
    elif n == 3:
        return 4
        # n이 3이라면 4를 출력한다.
    
    a= [0] * (n+1) # 0번째를 고려해서 n+1만큼 만든다.
    a[1]= 1; a[2]= 2; a[3]= 4 # 이거는 기본으로 저장해둔다.

    for i in range(4, n+1):
        a[i]= a[i-1] + a[i-2] + a[i-3]
        # n의 전 숫자, 전전 숫자, 전전전 숫자 총 3개를 더한다.
    return a[n]

t= int(input())
for i in range(t):
    n= int(input())
    print(sol(n))
```
