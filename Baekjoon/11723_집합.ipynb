### 3차 시도 : 성공

input().rstrip()을 생활화 합시다.

```python
from sys import stdin
input= stdin.readline

s= set()
for _ in range(int(input())):
    v = input().rstrip()

    if v == 'all':
        s= {i for i in range(1, 21)}
    elif v == 'empty':
        s= set()

    else:
        op, k = v.rstrip().split()
        if op == 'add':
            s.add(int(k))
        elif op == 'remove':
            if int(k) in s:
                s.remove(int(k))
        elif op == 'check':
            print(int(int(k) in s))
        else:
            s.remove(int(k)) if int(k) in s else s.add(int(k))
```
