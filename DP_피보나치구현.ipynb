### 1. DP - Memoization(하향식)

<pre>
한번 구한 계산 결과를 메모리 공간에 메모해두고 같은 식을 다시 호출하면 메모한 결과 그대로 가져오는 기법

코드 자세한 구현 방식은 아래 그림에 있음
</pre>

```python
def fibo_memo(x, memo):

    # 피보나치의 0번째 인덱스는 0, 1번째 인덱스는 1, 2번째 인덱스는 1이므로
    if x == 1 or x == 2:
        # return 1
        memo[x]= 1
        # 둘이 같은거 아닌가?
    
    # 이미 계산한 적이 있다면
    if memo[x] != 0:
        return memo[x]
    
    memo[x]= fibo_memo(x-1, memo) + fibo_memo(x-2, memo)

    return memo[x]
```

```python
# 100번째 피보나치 값은?
x= 100
memo= [0]*(x+1)
fibo_memo(x, memo)
```

### 2. DP - Tabulation(상향식)으로 구현

<pre>
하위 문제부터 천천히 해결하면서 더 큰 문제를 해결하는 기법을 말한다.
</pre>

```python
def fibo_tabul(x):
    tabul= [i for i in range(x+1)] 
    # 0번째 인덱스는 0, 1번재 인덱스는 1을 위해서 이렇게 초기화 한다.
    # 나머지는 for문 들어가면서 어차피 바뀜

    if x < 2: # x값이 0혹은 1일 경우 초기화 값을 리턴한다.
        return tabul[x]
    
    # 2번째 인덱스값부터 구현한다.
    for i in range(2, x+1):
        tabul[i]= tabul[i-1] + tabul[i-2]

    return tabul[x]
```

```python
# 100번째 피보나치 값은?
x= 100
fibo_tabul(x)
```
